```python
from __future__ import print_function 
import numpy as np 
import math
import pandas as pd
from sklearn import preprocessing


#기준정보 전처리

st_before_data = pd.read_csv('./data/MTS 기준정보(SAM2).CSV')

st_after_data = np.array(st_before_data[['축력 하중', '컬링 하중', '토크상한 누적시간', 'RPM 편차', '높이측정 - 완료위치']])

n, k = st_after_data.shape

for i in range(n):
    if st_after_data[i][0] > 1750:
        st_after_data[i][0] = 1750
    if st_after_data[i][2] < 0.023:
        st_after_data[i][2] = 0.023
    if st_after_data[i][3] < 46:
        st_after_data[i][3] = 46
st_result_data = np.vstack((np.array([1, 2, 3, 4, 5]), st_after_data))

np.savetxt('./pretreted_data/MTS 기준정보(SAM2)_전처리완료.CSV', st_result_data, fmt='%.5f', delimiter=' ')
print('기준정보 전처리 완료')

#분석Data 전처리

before_data = pd.read_csv('./data/SAM2 분석결과값(1월16일).CSV')

after_data = np.array(before_data[['축력 하중(kgf)', '컬링 하중(kgf)', '토크상한 누적시간', 'RPM 편차(rpm)', '높이측정 - 완료위치(mm)']])

n, k = after_data.shape

for i in range(n):
    if after_data[i][0] > 1750:
        after_data[i][0] = 1750
    if after_data[i][2] < 0.023:
        after_data[i][2] = 0.023
    if after_data[i][3] < 46:
        after_data[i][3] = 46
result_data = np.vstack((np.array([1, 2, 3, 4, 5]), after_data))

np.savetxt('./pretreted_data/SAM2 분석결과값(1월16일)_전처리완료.CSV', result_data, fmt='%.5f', delimiter=' ')
print('분석정보 전처리 완료')


#데이터 
OKdata=pd.read_csv('./pretreted_data/MTS 기준정보(SAM2)_전처리완료.CSV', delimiter=' ', encoding='utf-8' ) # SAM2데이터 from 익산1월
scaled_df_ok=(OKdata-OKdata.mean())/OKdata.std()
print('data \n', scaled_df_ok, '\n\n') # 확인 완료
k=OKdata.shape[1]
OKdata_corr=scaled_df_ok.corr() # 상관계수 구하기
print('상관계수(R)\n', OKdata_corr, '\n\n')
cor =np.linalg.inv(OKdata_corr) # 상관계수의 역함수
print('상관계수 역함수(R^-1)\n', cor, '\n\n')

#라인 데이터 검사
OKdata2=pd.read_csv('./pretreted_data/SAM2 분석결과값(1월16일)_전처리완료.CSV', delimiter=' ',encoding='utf-8' )
OKdata2=(OKdata2-OKdata.mean())/OKdata.std()
np_OKdata2=np.array(OKdata2)
j=np.arange(0,len(OKdata2),1)

#MD 값 계산
MD_result = []
for i in j :
    okdata_matrix=np_OKdata2[[i]]
    #print(NGdata_matrix)
    okdata_t=np_OKdata2[i].transpose()
    #print(NGdata_t)
    MD_squre1=np.dot( okdata_matrix,cor )/k
    # print(MD_squre1.shape)
    MD_squre2=np.dot( MD_squre1,okdata_t )
    MD_squre3=np.asscalar(MD_squre2)
    # MD_squre3_5=math.sqrt(MD_squre2)    D값
    MD_result.append(MD_squre3)

ok_mean=np.mean(MD_result)
print('MD 평균값 \n', ok_mean, '\n\n')
MD_result = np.array(MD_result)

#기존 Data에 추가
before_data['MD 결과값'] = MD_result

print('결과 Data\n',before_data, '\n\n\n')

dataframe = pd.DataFrame(before_data)
dataframe.to_csv('./result/SAM2 분석결과값.CSV')

print("== FINISHED ==")
```
